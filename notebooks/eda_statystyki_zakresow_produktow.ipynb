## 891 rows with missing values in Województwo

```python
df[
    df["Województwo"].isna()
    & (df["Powiat"].notna() | df["Nazwa świadczeniodawcy"].notna())
].shape
```

```python
df_powiat_woj = df_raw[["Województwo", "Powiat"]].dropna()

# Check for duplicates in the "Powiat" column
duplicated_powiat = df_powiat_woj[df_powiat_woj["Powiat"].duplicated(keep=False)]

# Get the unique "Powiat" values from the duplicated dataframe
duplicated_powiat_values = set(duplicated_powiat["Powiat"])

# Create dictionary with "Powiat" as keys and "Województwo" as values
powiat_to_wojewodztwo = dict(zip(df_powiat_woj["Powiat"], df_powiat_woj["Województwo"]))

# Identify rows where 'Województwo' is missing and 'Powiat' or 'Nazwa świadczeniodawcy' is not null
condition = df["Województwo"].isna() & (
    df["Powiat"].notna() | df["Nazwa świadczeniodawcy"].notna()
)

# Get the unique "Powiat" values from the duplicated dataframe
duplicated_powiat_values = set(duplicated_powiat["Powiat"])

# Filter rows where 'Powiat' is not in duplicated_powiat
filtered_df = df[condition & ~df["Powiat"].isin(duplicated_powiat_values)]

# Fill missing values in 'Województwo' using the mapping dictionary
df.loc[filtered_df.index, "Województwo"] = df.loc[filtered_df.index, "Powiat"].map(
    powiat_to_wojewodztwo
)
```
